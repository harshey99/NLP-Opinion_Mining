**Explain at least 3 steps that you will perform to preprocess the texts before training a classifier.**

Text Preprocessing is the process of cleaning textual data before it is given as input data to the model.  The raw data contains unwanted punctuation marks, numbers and alphabets in upper and lower case etc which can interfere during text analysis.

Data cleaning is defined as a procedure to add  missing data, remove unwanted and irrelevant data and discard duplicate values from the input dataset.

Hence, text pre-processing is done to ensure that all irregularities in the input data is removed.

> I have implemented the following text pre-processing techniques:

**1. Removal of all the Special Characters and Punctuation Marks**

>The special characters do not carry any important information. Therefore, I have  removed all the URLs, symbols etc., from the dataset for data cleaning. > Capital letters are used to indicate the begining of a new sentence. But the upper case letters can add noise while training the data.

> Stopwords are the most commonly occured words in English language. Stopwords mostly include "in", "is" etc,. among others. The stopwords have high frequency of occurences but do not contribute to Suggestion Detection. Therefore, it is important to remove them in order to condserve computing time for large volumes of data.

**5. Removal of Spaces**

> The inconsistent blank spaces are commonly found in any dataset and can lead to genearting inaccurate results during data analysis. Thus, it is imperative to remove any blank spaces from the dataset.

**Why is accuracy not the best measure for evaluating a classifier? Describe an evaluation metric**

Accuracy can be defined as the ratio of correct predictions to total number of predictions.
Accuracy is a single metric that can be used to summarise a model.
Accuracy works the best for a Balanced Classification  when there are equal number of samples in each class.
Accuracy cannot be used for a classification task whose distribution of class is skewed or on an imbalanced classification because of the intuition of a ML Practitioner.
ML Practitioners usually thaink that small rates of error and high accuracy scores such as close to 90 or above are ideal to evaluate the performance of model since they usuaaly work on small sized datasets with equal distribution of class.

Therefore, accuracy cannot be implemented on imbalanced class distribution.



**Which might work better than accuracy for a classification task such as suggestion detection.**

There are multiple evaluation metrics such as Accuracy, Precision, Recall nad F1 Score.

**Precision:**
It can be defined as the ratio of True Positives to the sum of True Positives and False Positives.
It does not consider True Negatives into consideration.
It is only best suited for Binary Classification tasks.
Comparison of the models cannot be done because it is not a single metrics.

**Recall:**
It can be defined as the ratio of True Positives to the sum of True Positives and False Negatives.
It does not make use of all the fields in the confusion matrix.
It focuses only on the Positive Class.
Comparison of the models cannot be done because it is not a single metrics.

**F-Score**

F-Score can be defined as the harmonic mean of Precision and Recall.
It is the ratio of product of Precision and Recall to the sum of Precision and Recall multiplied by 2.
The highest vale of F-Score is 1. It balances both Precision and Recall.
Since it is a single metrics it can be used to compare the models.
False Negatives can be minimized by maximizing precision and False Positives can be minimized by maximizing recall. Precision is given more significance.

**F Beta Score**

It prioritises either precision or accuarcy by adding additional weights in the form of Beta values.
The beta value is less than 1 when false positives are minimized whereas beta value is set to greater than 1 when false negatives are minimized. Recall is given more significance.

**Therefore, F-Score and F Beta Score are the most efficient evaluation metrics.**
